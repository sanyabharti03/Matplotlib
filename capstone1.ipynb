```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("C:\\Users\\KIIT0001\\Downloads\\student_data_sample.csv")
df
```

Line chart comparing Maths and Science marks against Roll No.

```python
plt.figure(figsize=(20,10))
plt.plot(df['Roll No'],df['Maths'],label="Maths",marker="D")
plt.plot(df['Roll No'],df['Science'],label="Science",marker=".",linestyle=':')
plt.grid()
plt.xlabel('Roll No')
plt.ylabel('Marks')
plt.legend(loc='best',fontsize='medium').get_frame().set_facecolor('blue')
# plt.tight_layout()
plt.title("MARKS OF STUDENTS")
plt.xticks(df['Roll No'][::5],rotation=45)
plt.savefig("Maths and Science Line Chart.png")
```

Subplots showing bar charts of Maths, Science, and English marks.

```python
fig=plt.figure()
ax=fig.subplots(3)
plt.grid(linestyle='--')
fig.suptitle("MARKS OF DIFF SUBJECTS")
ax[0].bar(df['Roll No'],df['Maths'])
ax[0].set_xlabel('Roll No')
ax[0].set_ylabel('Maths')
ax[0].set_title('Marks of Maths')
ax[0].set_ylim(0,100)

ax[1].bar(df['Roll No'],df['Science'])
ax[1].set_xlabel('Roll No')
ax[1].set_ylabel('Science')
ax[1].set_title('Marks of Science')
ax[1].set_ylim(0,100)

ax[2].bar(df['Roll No'],df['English'])
ax[2].set_xlabel('Roll No')
ax[2].set_ylabel('English')
ax[2].set_title('Marks of English')
ax[2].set_ylim(0,100)

fig.tight_layout()
plt.savefig("subplot.png")
```

Pie chart showing grade distribution for Maths.

```python

grade_array=[]
for i in range(0,100):
    if df.loc[i,'Maths']>90:
        grade='A'
        grade_array.append(grade)
    elif  df.loc[i,'Maths']>80:
        grade='B' 
        grade_array.append(grade)  
    elif df.loc[i,'Maths']>70:
        grade='C' 
        grade_array.append(grade)   
    elif df.loc[i,'Maths']>60:
        grade='D' 
        grade_array.append(grade)
    elif df.loc[i,'Maths']>50:
        grade='E'
        grade_array.append(grade)
    else:
        grade='F' 
        grade_array.append(grade)   
nparray=np.array(grade_array)        

df['Maths_grade']=grade_array 
explode=[0.2,0,0,0,0,0]   
count=df['Maths_grade'].value_counts().sort_index() 
plt.pie(count,labels=np.unique(nparray),autopct='%1.2f%%',shadow=True,explode=explode) 
plt.title("Maths Marks Distribution")
plt.axis('equal')
plt.savefig("Marks_Distribution.png")
        
```
